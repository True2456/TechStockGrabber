```python
#Format date for graphs
def dateConv(input):
    datetime = input
    datetime = pd.to_datetime(datetime)
    return datetime
```

```python
#Get Stock from ticker
class stockdata:
    def __init__(stock, ticker):
        stock.df = pdr.get_data_tiingo(ticker, api_key=key)
        stock.df.to_csv(ticker+'.csv')
        stock.df=pd.read_csv(ticker+'.csv')
        stock.datetime=dateConv(stock.df['date'])
        
```

```python
#Get stock data from ticker
##Plot data
DELL = stockdata("DELL")
HPE = stockdata("HPE")
AAPL = stockdata("AAPL")
plt.title('close price history')
plt.plot(DELL.datetime, DELL.df['close'], label = "Dell close")
plt.plot(HPE.datetime, HPE.df['close'], label = "HP close")
plt.plot(AAPL.datetime, AAPL.df['close'], label = "Apple close")
plt.xlabel('Date', fontsize=18)
plt.ylabel('close price USD ($)', fontsize=18)
plt.legend()
plt.show()
```

```python
plt.title('volume history')
plt.plot(DELL.datetime, DELL.df['volume'], label = "Dell volume")
plt.plot(HPE.datetime, HPE.df['volume'], label = "HP volume")
plt.plot(AAPL.datetime, AAPL.df['volume'], label = "Apple volume")
plt.xlabel('Date', fontsize=18)
plt.ylabel('Total volume', fontsize=18)
plt.legend()
plt.show()
```
